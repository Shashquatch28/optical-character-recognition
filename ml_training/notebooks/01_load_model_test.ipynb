```python
# Import the necessary classes from the transformers library
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
import torch
```

```python
# 1. ----- Define Model -----
MODEL_NAME = "microsoft/trocr-base-handwritten"

print(f"Attempting to load processor for: {MODEL_NAME}")

# 2. ----- Load the TrOCRProcessor -----
try:
    processor = TrOCRProcessor.from_pretrained(MODEL_NAME)
    print("...Processor loaded successfully!")

    # 3.----- Load the TrOCR Model -----
    print(f"Attempting to load model: {MODEL_NAME}")
    model = VisionEncoderDecoderModel.from_pretrained(MODEL_NAME)
    print("...Model loaded successfully!")

    # 4. ----- Check GPU Availability -----
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.to(device)
    
    print("\n--- Setup Complete ---")
    print(f"Model and processor are loaded and running on: {device}")

except Exception as e:
    print(f"\n--- ERROR ---")
    print(f"An error occurred: {e}")
    print("Please check your internet connection and if the library 'transformers' is installed correctly.")
```
